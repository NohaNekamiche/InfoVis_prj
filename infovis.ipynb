pour le premier cas d'etude : **Avec 4 variables (sexe,age,fumer ,alcoolique)**

nous avons etudier la distribution des hommes et femmes fummeur et alcoolique selon l'age

```python
import pandas as p
df=p.read_csv('pretraitement_csv.csv')
df.drop(['Unnamed: 0','Unnamed: 0.1'],1,inplace=True)
df.head()
```

faire des réquets sur la table des données pour partitionnée nos données en :

1.   Homme fumeur et alcoolique 
2.   Homme fumeur et Non alcoolique 
3.   Homme Non fumeur et Non alcoolique 
4.   Femme fumeur et  alcoolique 
5.   Homme fumeur et Non alcoolique 
6.   Homme Non fumeur et Non alcoolique 



```python

homme_fume_alcoo=df.loc[(df['sexe']=='HOMME')&(df['fume']=='OUI')&(df['alcoo']=='OUI')]
homme_fume_non_alcoo=df.loc[(df['sexe']=='HOMME')&(df['fume']=='OUI')&(df['alcoo']=='NON')]
homme_non_fume_non_alcoo=df.loc[(df['sexe']=='HOMME')&(df['fume']=='NON')&(df['alcoo']=='NON')]
femme_fume_alcoo=df.loc[(df['sexe']=='FEMME')&(df['fume']=='OUI')&(df['alcoo']=='OUI')]
femme_fume_non_alcoo=df.loc[(df['sexe']=='FEMME')&(df['fume']=='OUI')&(df['alcoo']=='NON')]
femme_non_fume_non_alcoo=df.loc[(df['sexe']=='FEMME')&(df['fume']=='NON')&(df['alcoo']=='NON')]
```

creer une nouvelle table qui contient les données filtre au-dessous

```python
min_age=min(df['age'])
max_age=max(df['age'])

da={'age':[],
   'homme_fum_alcoo':[],
   'homme_fum_non_alcoo':[],
    'homme_non_fume_non_alcoo':[],
   'femme_fum_alcoo':[],
   'femme_fum_non_alcoo':[],
   'femme_non_fum_non_alcoo':[]}
data_f=p.DataFrame(da)
h_f_a=0
h_f_na=0
h_nf_na=0
f_f_a=0
f_f_na=0
f_nf_na=0

i=min_age
while (i<=max_age):
    h_f_a=sum(homme_fume_alcoo.age==i)
    h_f_na=sum(homme_fume_non_alcoo.age==i)
    h_nf_na=sum(homme_non_fume_non_alcoo.age==i)
    f_f_a=sum(femme_fume_alcoo.age==i)
    f_f_na=sum(femme_fume_non_alcoo.age==i)
    f_nf_na=sum(femme_non_fume_non_alcoo.age==i)
    data_f.loc[data_f.shape[0]]=[i,h_f_a,h_f_na,h_nf_na,f_f_a,f_f_na,f_nf_na]
    i+=1
    
data_f
    
    
```

sauvegarder les resultats dans un fichier csv & excel

```python
data_f.to_excel('filter_data.xlsx')
data_f.to_csv('filter_data.csv', index=False)
```
